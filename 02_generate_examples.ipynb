```julia
using PanGraph;
using Random;

Random.seed!(123456);

sequences=[PanGraph.Simulation.randseq(300) for _ in 1:3]

open("tests/data/randseqs.fa", "w") do io
  for (i, sequence) in enumerate(sequences)
    PanGraph.Simulation.write_fasta(io, 'A' + i - 1, sequence)
  end
end
```
