```julia
using PanGraph;
using Random;

Random.seed!(123456);

sequences=[PanGraph.Simulation.randseq(200) for _ in 1:10]

TESTDIR="example_datasets"

open("$TESTDIR/smol_01.fa", "w") do io
  for (i, sequence) in enumerate(sequences)
    PanGraph.Simulation.write_fasta(io, 'A' + i - 1, sequence)
  end
end
```
