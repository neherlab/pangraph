```julia
compare = Graphs.Mash.distance;
tree = Align.ordering(compare, graphs...);
tree
```

```julia
tree = Align.balance(tree)
tree
```

```julia
tips = Dict(collect(keys(G.sequence))[1] => G for G in graphs);
tips
```

```julia
aligner(contigs₁, contigs₂) = Minimap.align(contigs₁, contigs₂, minblock, sensitivity);
minblock = 100;
sensitivity = 10;
```

```julia
G = nothing
for (n_clade, clade) ∈ enumerate(Align.postorder(deepcopy(tree)))
    if Align.isleaf(clade)
        close(clade.graph)
        msg = (tips[clade.name], n_clade)
        put!(clade.parent.graph, msg)
        # JSON.print(msg, 2)
    else
        Gₗ, Pₗ = take!(clade.graph)
        Gᵣ, Pᵣ = take!(clade.graph)
        close(clade.graph)
        # JSON.print(take!(clade.graph), 2)

        G₀ = Graphs.copy_graph(Gₗ) # TODO

        if clade.parent !== nothing
            msg = (G₀, n_clade)
            put!(clade.parent.graph, msg)
        else
            G = G₀
        end
    end
end
JSON.print(G, 2)
```
